# Conclusion
There are multiple problems with this version of the Pf6.2 manifest from fits:
- A small number of lanelets have multiple rows, with one version containing "no_cal" in the file path
- Most lanelets have no study associated. This is required for all rows
- There are many lanelets from studies which should have been excluded, e.g. 1176-PF-KE-CP1, 1204-PF-GM-CP1 and R&D studies
- Most worryingly, there are a large number of lanelets that are missing but that were included in previous builds. In Richard's version of the manifest most, but not all of these were missing sanger_sample_id. I couldn't spot any other obvious pattern.
